```python
import math
```

```python
import numpy as np
```

```python
from sklearn import cross_validation
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
data = np.loadtxt("q1_data.csv", delimiter = ",")
```

```python
clf = LogisticRegression()
```

```python
nData =[]
```

```python
nClf =[]
```

```python
print(len(data))
```

```python
for i in range(len(data)):
    nData.append([math.sqrt(data[i,0] * data[i, 0] + data[i, 1] * data[i, 1])])
    nClf.append(data[i,2])
```

```python
nData
```

```python
scores = cross_validation.cross_val_score(clf, nData, nClf, cv=10)
```

```python
scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
